```python
import requests

responce = requests.get('https://stackoverflow.com/questions/415511/how-do-i-get-the-current-time-in-python', headers ={"User-Agent":"Mozilla/5.0"})
```

```python
responce
```

```python
from bs4 import BeautifulSoup
soup = BeautifulSoup(responce.content,'html.parser')
print(soup.prettify())
```

```python
question_title = soup.find('title')

print(question_title.text)
```

```python
question_title = soup.find('a', class_="question-hyperlink")
print("Question title -",question_title.text)
```

```python

answers = soup.find_all('div', class_='answer')
for answer in answers:
    answer_text = answer.find('div', class_='s-prose')
    print("Answer:", answer_text)
```
